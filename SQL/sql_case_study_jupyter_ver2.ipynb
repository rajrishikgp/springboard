```python
#10. Produce a list of facilities with a total revenue less than 1000 grouped by month
import sqlite3
import pandas as pd
conn = sqlite3.connect('C:/Users/rsaha/Anaconda3/SQL/sqlite_db_pythonsqlite.db')
query = '''SELECT f.name, (SUM(f.membercost * b.slots) + SUM(f.guestcost * b.slots)) as Revenue, strftime('%m', b.starttime) as 'Month'
FROM Facilities as f
INNER JOIN Bookings as b
ON f.facid = b.facid
GROUP BY f.name, Month
HAVING Revenue < 1000
ORDER BY Revenue'''

df = pd.read_sql_query(query, conn)
df
```

```python
#Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order
import sqlite3
import pandas as pd
conn = sqlite3.connect('C:/Users/rsaha/Anaconda3/SQL/sqlite_db_pythonsqlite.db')
query = '''SELECT m1.surname, m1.firstname, m2.surname as recommendersurname, m2.firstname as recommenderfirstname
FROM Members as m1
LEFT OUTER JOIN Members as m2
ON m1.recommendedby = m2.memid
WHERE m1.recommendedby != 0
ORDER BY m1.surname, m1.firstname'''

df = pd.read_sql_query(query, conn)
df
```

```python
#Q12: Find the facilities with their usage by member, but not guests
import sqlite3
import pandas as pd
conn = sqlite3.connect('C:/Users/rsaha/Anaconda3/SQL/sqlite_db_pythonsqlite.db')
query = '''SELECT f.name, m.firstname, m.surname, SUM(b.slots) as 'usage'
FROM Members as m
LEFT JOIN Bookings as b
ON m.memid = b.memid
LEFT JOIN Facilities as f
on f.facid = b.facid
WHERE (m.firstname NOT LIKE "%GUEST%") & (f.name IS NOT NULL) 
GROUP BY f.name, m.surname, m.firstname'''

df = pd.read_sql_query(query, conn)
df
```

```python
#Q13: Find the facilities usage by month, but not guests
import sqlite3
import pandas as pd
conn = sqlite3.connect('C:/Users/rsaha/Anaconda3/SQL/sqlite_db_pythonsqlite.db')
query = '''SELECT f.name, SUM(b.slots) as 'usage', strftime('%m', b.starttime) as 'Month'
FROM Members as m
LEFT JOIN Bookings as b
ON m.memid = b.memid
LEFT JOIN Facilities as f
on f.facid = b.facid
WHERE (m.firstname NOT LIKE '%GUEST%') & (strftime('%m', b.starttime) IS NOT NULL)
GROUP BY f.name, Month
'''

df = pd.read_sql_query(query, conn)
df
```
