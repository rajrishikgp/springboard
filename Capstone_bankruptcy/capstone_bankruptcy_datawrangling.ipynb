```python
df.columns = [c.replace(' ', '_') for c in df.columns]
```

```python
df.head()
```

```python
sns.histplot(x=df._Operating_Expense_Rate, data= df)
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df[['_Operating_Expense_Rate']])
df['normalized_operating_expense_rate'] = scaler.transform(df[['_Operating_Expense_Rate']])
df.drop(columns= '_Operating_Expense_Rate', inplace=True)
df.head()
```

```python
sns.histplot(x=df._Research_and_development_expense_rate, data= df)
plt.show()
```

```python
scaler.fit(df[['_Research_and_development_expense_rate']])
df['normalized_Research_and_development_expense_rate'] = scaler.transform(df[['_Research_and_development_expense_rate']])
df.drop(columns= '_Research_and_development_expense_rate', inplace=True)
df.head()
```

```python
sns.histplot(x=df._Total_Asset_Growth_Rate, data=df)
plt.show()
```

```python
scaler.fit(df[['_Total_Asset_Growth_Rate']])
df['normalized_Total_Asset_Growth_Rate'] = scaler.transform(df[['_Total_Asset_Growth_Rate']])
df.drop(columns= '_Total_Asset_Growth_Rate', inplace=True)
df.head()
```

```python
df = df.rename(columns={'_Inventory_Turnover_Rate_(times)': 'Inventory_Turnover_Rate'})
```

```python
sns.histplot(x=df.Inventory_Turnover_Rate, data=df)
plt.show()
```

```python
scaler.fit(df[['Inventory_Turnover_Rate']])
df['normalized_Inventory_Turnover_Rate'] = scaler.transform(df[['Inventory_Turnover_Rate']])
df.drop(columns= 'Inventory_Turnover_Rate', inplace=True)
df.head()
```
